```python
from transformers import pipeline
import pandas as pd
from datasets import load_dataset

# Load translation pipeline for Sinhala to English
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-inc-en")

# Load the Divaina dataset from Hugging Face
ds = load_dataset("Pamzyy/Partially_uploaded_divaina")

# Convert the dataset to a pandas DataFrame for easier manipulation
df = pd.DataFrame(ds['train'])

# Translate each text entry from Sinhala to English in the 'content' column
df['Translated'] = df[' content'].apply(lambda x: translator(x)[0]['translation_text'])

# Save the DataFrame with the translations to a CSV file
df.to_csv("translated_divaina_dataset.csv", index=False)

print("Translation completed and saved to translated_divaina_dataset.csv")
```
