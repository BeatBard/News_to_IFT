```python
import pandas as pd
from deep_translator import GoogleTranslator
import time
```

```python
# Initialize the Google Translator with deep_translator
translator = GoogleTranslator(source='auto', target='si')

# Load the original CSV file
file_path = 'instructions_content.csv'
df = pd.read_csv(file_path)

# Ensure the column names are correct
instructions_column = 'Generated Instructions'
original_content_column = 'Original Content'
```

```python
translated_instructions = []
```

```python
for instruction in df[instructions_column]:
    if pd.notna(instruction):  # Ensure the instruction is not NaN
        try:
            translated_text = translator.translate(instruction)
            translated_instructions.append(translated_text)
        except Exception as e:
            print(f"Error translating '{instruction}': {e}")
            translated_instructions.append('')  # Handle errors gracefully
        # Add a delay to prevent overloading the API
        time.sleep(1)  # 1-second delay between API requests
    else:
        translated_instructions.append('')  # Handle NaN values

# Add the translated instructions to the dataframe
df['Translated Instructions'] = translated_instructions

# Create a new dataframe with only 'Translated Instructions' and 'Original Content' columns
new_df = df[['Translated Instructions', original_content_column]]

# Save the new dataframe to a CSV file
new_file_path = 'translated_instruction_dataset.csv'
new_df.to_csv(new_file_path, index=False)

print(f'Translated dataset saved to {new_file_path}')
```
